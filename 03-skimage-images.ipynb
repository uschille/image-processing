```python
import numpy as np
import skimage
import imageio.v3 as iio
import matplotlib.pyplot as plt
```

```python
image = skimage.io.imread("data/sudoku.png")
gray_image = skimage.color.rgb2gray(image)
gray_image = skimage.util.img_as_ubyte(gray_image)
image.shape, image.dtype, gray_image.shape, gray_image.dtype
```

```python
image = iio.imread(uri="data/sudoku.png", mode="L")
image = np.array(image)

plt.imshow(image, cmap="gray")

image.shape, image.dtype, image.flags
```

```python
image[image > 192] = 192

plt.imshow(image, cmap="gray", vmin=0, vmax=255);
```
